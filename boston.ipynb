### <font color=deeppink> Importing libraries </font>

```python
import os
import sys
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
import matplotlib.pyplot as plt
```

### <font color=deeppink> Loading Boston Housing Dataset </font>

```python
from tensorflow.keras.datasets import boston_housing
(train_data, train_labels), (test_data, test_labels) = boston_housing.load_data()
```

```python
print(train_data.shape)
print(test_data.shape)
```

### <font color=deeppink> Data pre processing </font>

```python
mean = train_data.mean(axis=0) 
print('Mean =', mean)
train_data -= mean

std_dev = train_data.std(axis=0)
print('Std Dev = ', std_dev)
train_data /= std_dev

test_data -= mean
test_data /= std_dev
```

## <font color=deeppink> Model Building </font>

Creating sequencial model with 3 dense layer only.

```python
model0 = keras.Sequential([ 
        tf.keras.layers.Dense(32, activation='relu', name ='Dense_Layer_1',input_shape=(13,)),
        tf.keras.layers.Dense(64, activation='relu',name ='Dense_Layer_2'),
        tf.keras.layers.Dense(1,name ='Dense_Layer_3')
    ])
print("--Model Building Successful--")
```

```python
model0.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5),
              loss='mse', 
              metrics=['mae'], ) 
print("--Model compilation successful--")
```

```python
history0 = model0.fit(train_data, train_labels, epochs=40, batch_size=16)
```

```python
model0.summary()
```

```python
[mse, mae] = model0.evaluate(test_data, test_labels) 
```

```python
ypred = model0.predict(test_data)
print("--Testing Prices--")
print('Actual Price = {} \nPredicted Price = {}'.format(test_labels[1], ypred[1]))
```

```python
history_dict0 = history0.history
loss_value0 = history_dict0['loss']
mae_value0 = history_dict0['mae']
epochs0 = range(1, len(loss_value0) + 1)
```

 ## <font color=deeppink> Model building with Batch normalization </font>

Creating sequential layer with 3 dense layer and 1 Batch Normalization layer.

```python
model1 = keras.Sequential([ 
        tf.keras.layers.Dense(32, activation='relu', name ='Dense_Layer_1',input_shape=(13,)),
        tf.keras.layers.Dense(64, activation='relu',name ='Dense_Layer_2'),
        tf.keras.layers.BatchNormalization(name ='Batch_Normalization_Layer_1'),
        tf.keras.layers.Dense(1,name ='Dense_Layer_3')
    ])
print("--Model Building Successful--")
```

### <font color=deeppink> Compiling the model-1 </font>

```python
model1.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5),
              loss='mse', 
              metrics=['mae'], ) 
print("--Model compilation successful--")
```

### <font color=deeppink> Model-1 Training </font>

```python
history1 = model1.fit(train_data, train_labels, epochs=40, batch_size=16)
```

Loss and Mean Absulate Error are to high

### <font color=deeppink> Model-1 Summary </font>

```python
model1.summary()
```

```python
[mse, mae] = model1.evaluate(test_data, test_labels) 
```

```python
ypred = model1.predict(test_data)
print("--Testing Prices--")
print('Actual Price = {} \nPredicted Price = {}'.format(test_labels[1], ypred[1]))
```

```python
history_dict1 = history1.history
loss_value1 = history_dict1['loss']
mae_value1 = history_dict1['mae']
epochs1 = range(1, len(loss_value1) + 1)
```

## <font color=deeppink>Adding L2 weight regularization to the model</font>

```python
from tensorflow.keras import regularizers
model2 = keras.Sequential([
 tf.keras.layers.Dense(32, kernel_regularizer=regularizers.l2(0.002), activation="relu",input_shape=(13,)),
 tf.keras.layers.Dense(64, kernel_regularizer=regularizers.l2(0.002), activation="relu"),
 tf.keras.layers.BatchNormalization(name ='Batch_Normalization_Layer_1'),
 tf.keras.layers.Dense(1,kernel_regularizer=regularizers.l2(0.002))
])
print("--Model Successfull--")
```

### <font color=deeppink>Model-2 Compiling</font>

```python
model2.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5),
              loss='mse', 
              metrics=['mae'], ) 
print("--Model compilation successful--")
```

### <font color=deeppink>Model-2 Training</font>

```python
history2 = model2.fit(train_data, train_labels, epochs=40, batch_size=16)
```

### <font color=deeppink>Model-2 Summary</font>

```python
model2.summary()
```

### <font color=deeppink>Model-2 Evaluation</font>

```python
[mse, mae] = model2.evaluate(test_data, test_labels) 
```

```python
ypred = model2.predict(test_data)
print("--Testing Prices--")
print('Actual Price = {} \nPredicted Price = {}'.format(test_labels[1], ypred[1]))
```

```python
history_dict2 = history2.history

loss_value2 = history_dict2['loss']
mae_value2 = history_dict2['mae']
epochs2 = range(1, len(loss_value2) + 1)
```

### <font color=deeppink> Plotting Batch normalization Vs L2 Regularization</font>

```python
plt.plot(epochs0, loss_value0, 'g', label='Training Loss')
plt.plot(epochs1, loss_value1, 'b', label='Training Loss of after Batch Normalization')
plt.plot(epochs2, loss_value2, 'r', label='Training Loss after applying L2 Regularization')
plt.title('Training Loss with Batch normalization and L2 Regularization')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

plt.plot(epochs0, mae_value0, 'g', label='MAE Loss')
plt.plot(epochs1, mae_value1, 'b', label='MAE Loss of after Batch Normalization')
plt.plot(epochs2, mae_value2, 'r', label='MAE Loss after applying L2 Regularization')
plt.title('MAE Loss with Batch normalization and L2 Regularization')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### <font color=deeppink> Calculating the Bias and Variance</font>

```python
from mlxtend.evaluate import bias_variance_decomp
mse, bias, var = bias_variance_decomp(model2, 
                                      train_data, train_labels, 
                                      test_data, test_labels, 
                                      loss='mse', 
                                      num_rounds=200, 
                                      random_seed=1)
# summarize results
print('MSE: %.3f' % mse)
print('Bias: %.3f' % bias)
print('Variance: %.3f' % var)
```

We have a high bias and a low variance!!
